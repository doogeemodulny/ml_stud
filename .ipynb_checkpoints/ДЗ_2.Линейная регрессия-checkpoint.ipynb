### –ë—É–¥–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å —Ü–µ–Ω—É –¥–æ–º–∞ üìù

–í–ê–ñ–ù–û: –ü—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –Ω–µ –∑–∞–±—ã–≤–∞–π—Ç–µ —É–∫–∞–∑–∞—Ç—å –Ω–∞–∑–≤–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∏ –æ–±–æ–∑–Ω–∞—á–∏—Ç—å, —á—Ç–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–æ –ø–æ –∫–∞–∂–¥–æ–π –∏–∑ –æ—Å–µ–π

–í –±–ª–æ–∫–µ –Ω–∏–∂–µ –∏–º–ø–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –≤—Å–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –≤–∞–º –ø–æ–Ω–∞–¥–æ–±—è—Ç—Å—è –¥–ª—è –∑–∞–ø—É—Å–∫–∞ –Ω–æ—É—Ç–±—É–∫–∞

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# some lib
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
```

**–°–∫–∞—á–∞–π—Ç–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ –ø–æ —Å—Å—ã–ª–∫–µ: https://www.kaggle.com/c/house-prices-advanced-regression-techniques**

```python
data = pd.read_csv('train_hw2.csv', header=0)
```

‚úÖ –ï—Å–ª–∏ –≤ –¥–∞–Ω–Ω—ã—Ö –±–µ—Å—Å–º—ã—Å–ª–µ–Ω–Ω—ã–µ —Å—Ç–æ–ª–±—Ü—ã? –ï—Å–ª–∏ –¥–∞, –∏–∑–±–∞–≤—å—Ç–µ—Å—å –æ—Ç –Ω–∏—Ö –∏ –æ–±—ä—è—Å–Ω–∏—Ç–µ, –ø–æ—á–µ–º—É —Ç–∞–∫ –ø–æ—Å—Ç—É–ø–∏–ª–∏.

```python
data_2 = data.iloc[:,1:] # –£–¥–∞–ª—è—é ID –ø–æ—Ç–æ—Ç–º—É —á—Ç–æ —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ –ø–æ—Ä—è–¥–∫–æ–≤—ã–π –Ω–æ–º–µ—Ä —Å—Ç—Ä–æ–∫–∏
data_2.head()
```

‚úÖ –ï—Å—Ç—å –ª–∏ –≤ –¥–∞–Ω–Ω—ã—Ö –ø—Ä–æ–ø—É—Å–∫–∏? –ï—Å–ª–∏ –¥–∞, —Ç–æ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞, –≤ –∫–æ—Ç–æ—Ä–æ–º –æ–Ω–∏ –∏–º–µ—é—Ç—Å—è, –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ –∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏ –∏—Ö –¥–æ–ª—é –æ—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ –∑–Ω–∞—á–µ–Ω–∏–π. –ß—Ç–æ –≤—ã –Ω–∞–±–ª—é–¥–∞–µ—Ç–µ?

```python
null_percentage = data_2.isnull().mean()
null_count = data_2.isnull().sum()

nulls_stats = pd.DataFrame({
    "Feature": data_2.columns,
    "Nulls count": null_count.values,
    "Nulls %": null_percentage.values
})

nulls_stats.values
```

‚úÖ –ò–∑–±–∞–≤—å—Ç–µ—Å—å –æ—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤. –î–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –ø—Ä–∏–º–µ–Ω–µ–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –æ–±–æ—Å–Ω—É–π—Ç–µ —Å–≤–æ–µ —Ä–µ—à–µ–Ω–∏–µ. 

```python
threshold = 0.3
data_3 = data_2.loc[:, null_percentage<threshold] # —É–±–∏—Ä–∞–µ–º —Å—Ç–æ–±—Ü—ã –≥–¥–µ –¥–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ –±–æ–ª—å—à–µ 30%
data_3.shape
```

```python
numerics = data_3.select_dtypes(exclude=["object"])
categoricals = data_3.select_dtypes(include=["object"])
print(numerics.columns.values, categoricals.columns.values)
numerics.shape, categoricals.shape
```

```python
numerics.fillna(numerics.median(), inplace=True)
categoricals.fillna(categoricals.mode().iloc[0], inplace=True)
df = pd.concat([numerics, categoricals], axis=1)
df.head()
```

‚úÖ –ü—Ä–æ–≤–µ—Ä—å—Ç–µ, —á—Ç–æ –≤—ã –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –∏–∑–±–∞–≤–∏–ª–∏—Å—å –æ—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤

```python
df.info()
```

‚úÖ –û–±—Ä–∞–±–æ—Ç–∞–π—Ç–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏. –û–±—ä—è—Å–Ω–∏—Ç–µ –≤—ã–±–æ—Ä –º–µ—Ç–æ–¥–∞ (one-hot-encoding, label encoding, ...)

```python
categoricals_ohe = pd.get_dummies(categoricals, drop_first=True, dtype=np.int64) 
#One-Hot-Encoding –ø–æ—á—Ç–∏ —É–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω—ã–π —Å–ø–æ—Å–æ–± –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è, –µ–≥–æ –æ—Å–Ω–æ–≤–Ω–æ–π –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ–∫ - –±–æ–ª—å—à–æ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ø–∞–º—è—Ç–∏. –ù–æ —É –Ω–∞—Å –º–∞–ª–æ –¥–∞–Ω–Ω—ã—Ö —Ç–∞–∫ —á—Ç–æ –û–ö
categoricals_ohe.head()
```

```python
df = pd.concat([numerics, categoricals_ohe], axis=1)
```

‚úÖ –í—ã—á–∏—Å–ª–∏—Ç–µ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ –ø–æ–ø–∞—Ä–Ω—É—é –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –ü–∏—Ä—Å–æ–Ω–∞ –º–µ–∂–¥—É –≤—Å–µ–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏. –ö–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å?

```python
df.corr()
```

```python
corr = numerics.corr()
sns.heatmap(corr)
```

‚úÖ –ù–∞–π–¥–∏—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –∞–±—Å–æ–ª—é—Ç–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞ —Å –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º. –ò–∑–æ–±—Ä–∞–∑–∏—Ç–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –Ω–∞–π–¥–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –æ—Ç –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è.

```python
min_max_corr = df.corr()['SalePrice'].sort_values(ascending=False)
df.corr()['SalePrice'].sort_values(ascending=False)
```

```python
target = numerics["SalePrice"]
numerics = numerics.iloc[:, :-1]
df.drop("SalePrice", axis=1, inplace=True)
df.shape, numerics.shape
```

‚úÖ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è. –î–ª—è –∏–∑–±–∞–≤–ª–µ–Ω–∏—è –æ—Ç —Ä–∞–∑–Ω–∏—Ü—ã –≤ –º–∞—Å—à—Ç–∞–±–∞—Ö, –∞ —Ç–∞–∫–∂–µ "—Å–º–µ—â–µ–Ω–∏—è" —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –≤ —Å—Ç–æ—Ä–æ–Ω—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ, –º–æ–∂–Ω–æ –ø—Ä–æ–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞—Ç—å –µ–µ (—ç—Ç–æ –æ–±—Ä–∞—Ç–∏–º–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –ø–æ—ç—Ç–æ–º—É —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –ª–µ–≥–∫–æ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å). 

–ü–æ–¥—Å–∫–∞–∑–∫–∞: –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å `numpy.log1p`, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ $y \to \ln\left(1 + y\right)$. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Ç –Ω–æ–≤–æ–≥–æ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è. –û–ø–∏—à–∏—Ç–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è.

–í–ê–ñ–ù–û: –¥–∞–ª–µ–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–ª—É—á–∏–ª—Å—è –ø–æ—Å–ª–µ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏—è.

```python
sns.histplot(target)
```

```python
logtarget = np.log1p(target)
sns.histplot(logtarget)
```

‚úÖ –†–∞–∑–±–µ–π—Ç–µ –≤—ã–±–æ—Ä–∫—É –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤ –æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 75:25. –ó–∞—Ñ–∏–∫—Å–∏—Ä—É–π—Ç–µ –ø—Ä–∏ —Ä–∞–∑–±–∏–µ–Ω–∏–∏ –∑–Ω–∞—á–µ–Ω–∏–µ `random_state` = 13. –ù–µ –∑–∞–±—ã–≤–∞–π—Ç–µ –ø—Ä–æ–ª–æ–≥–æ—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞—Ç—å —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é (–∫–∞–∫ –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º –∑–∞–¥–∞–Ω–∏–∏)

```python
X_train, X_test, y_train, y_test = train_test_split(df, logtarget, random_state=13)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

‚úÖ –ü—Ä–∏–º–µ–Ω–∏—Ç–µ –∫ –¥–∞–Ω–Ω—ã–º 

- –õ–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é
- Lasso
- Ridge

–î–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –º–µ—Ç–æ–¥–æ–≤ –ø–æ–¥–±–µ—Ä–∏—Ç–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Å –ø–æ–º–æ—â—å—é –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏. –û–±—É—á–∏—Ç–µ –∞–ª–≥–æ—Ä–∏—Ç–º—ã —Å –ª—É—á—à–∏–º–∏ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –∏ –æ—Ü–µ–Ω–∏—Ç–µ –∫–∞—á–µ—Å—Ç–≤–æ –ø–æ –º–µ—Ç—Ä–∏–∫–∞–º 
- RMSE 
- MAE
- $R^2$

–û–ø–∏—à–∏—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

```python
linreg = LinearRegression()
lasso = Lasso()
ridge = Ridge()
```

```python
linreg = linreg.fit(X_train, y_train) # —Ç–∞–º –Ω–µ—á–µ–≥–æ –ø–æ–¥–±–∏—Ä–∞—Ç—å –ø–æ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏
print("RMSE:", root_mean_squared_error(linreg.predict(X_test), y_test))
print("MAE:", mean_absolute_error(linreg.predict(X_test), y_test))
print("R2:", r2_score(linreg.predict(X_test), y_test))
```

```python
parameters = {'alpha':np.logspace(-4, 2)}
lasso_cv = GridSearchCV(lasso, parameters).fit(X_train, y_train)
lasso_cv.best_params_
```

```python
print("RMSE:", root_mean_squared_error(lasso_cv.predict(X_test), y_test))
print("MAE:", mean_absolute_error(lasso_cv.predict(X_test), y_test))
print("R2:", r2_score(lasso_cv.predict(X_test), y_test))
```

```python
parameters = {'alpha':np.logspace(-4, 2)}
ridge_cv = GridSearchCV(ridge, parameters).fit(X_train, y_train)
ridge_cv.best_params_
```

```python
print("RMSE:", root_mean_squared_error(ridge_cv.predict(X_test), y_test))
print("MAE:", mean_absolute_error(ridge_cv.predict(X_test), y_test))
print("R2:", r2_score(ridge_cv.predict(X_test), y_test))
```

‚úÖ –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∑–Ω–∞—á–µ–Ω–∏–π –≤–µ—Å–æ–≤ –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, Lasso –∏ Ridge. –û–ø–∏—à–∏—Ç–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è, –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

```python
sns.histplot(linreg.coef_)
plt.title("Linear Regression weights")
```

```python
best_lasso = lasso_cv.best_estimator_
sns.histplot(best_lasso.coef_, bins=20)
plt.title("Lasso Regression weights")
```

```python
best_ridge = ridge_cv.best_estimator_
sns.histplot(best_ridge.coef_)
plt.title("Ridge Regression weights")
```
