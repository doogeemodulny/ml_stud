## Анализ и предобработка данных

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
X_train.shape, X_test.shape
```

## Добавление категориальных признаков

Добавим к предобработанным вещественным признакам категориальные. Но для начала их тоже необходимо предобработать.

Преобразуйте все категориальные признаки с помощью метода one-hot-encoding (например, это можно сделать с помощью функции [pandas.get_dummies](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.get_dummies.html) или [DictVectorizer](http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.DictVectorizer.html) / [OneHotEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) из sklearn).

После преобразования категориальных признаков, добавьте их к вещественным предобработанным данным для обучения.

>PS Напоминаю, что важно удалить колонку с целевым значением

```python
X_train[categoricals].head()
```

```python
X_train = pd.get_dummies(X_train, columns=categoricals).astype(np.float64)
X_test = pd.get_dummies(X_test, columns=categoricals).astype(np.float64)
X_train.head()
```

## Обучение классификаторов

В начале посмотрим, как работает подбор параметров по сетке и как влияет на качество разбиение выборки. Сейчас и далее будем рассматривать 2 алгоритма:
 - [kNN](http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)
 - [DecisonTree](http://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier)

Для начала выберем один гиперпараметр, который будем оптимизировать:
 - kNN — число соседей (*n_neighbors*)
 - DecisonTree — глубина дерева (*max_depth*)
 
Остальные параметры оставляйте в значениях по умолчанию.

Для каждого алгоритма подберите оптимальные значения указанных гиперпараметров. Постройте график среднего значения качества по кросс-валидации алгоритма при заданном значении гиперпараметра, на котором также отобразите доверительный интервал.

Что вы можете сказать о получившихся графиках? Какой алгоритм справился лучше? Какой быстрее? Почему?

```python
# your code here
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import cross_val_score
```

```python
accuracy = make_scorer(accuracy_score)

confidence_intervals = []
knn_results = [[],[]]
for n in tqdm(range(1, 100, 2)):
    neigh = KNeighborsClassifier(n_neighbors=n)
    knn_results[0].append(n)
    cv_scores = cross_val_score(neigh, X_train, y_train, cv=10, scoring=accuracy)
    knn_results[1].append(cv_scores.mean())
    std_err = cv_scores.std() / np.sqrt(len(cv_scores))
    confidence_interval = 1.96 * std_err  # 95% доверительный интервал
    confidence_intervals.append(confidence_interval)

knn_results[0] = np.array(knn_results[0])
knn_results[1] = np.array(knn_results[1])
confidence_intervals = np.array(confidence_intervals)

plt.figure(figsize=(10, 6))
plt.plot(knn_results[0], knn_results[1], label='Mean accuracy', marker='+', color='green')
plt.fill_between(knn_results[0], 
                 knn_results[1] - confidence_intervals, 
                 knn_results[1] + confidence_intervals, 
                 color='green', alpha=0.2, label='95% confidence interval')
plt.xlabel("Number of neighbors")
plt.ylabel("Accuracy on cross-validation")
plt.xticks(np.arange(1, 100, 4))
plt.title("Accuracy on KNN")
plt.legend()
plt.grid()
plt.show()
```

```python
accuracy = make_scorer(accuracy_score)

confidence_intervals = []
dt_results = [[],[]]
for d in tqdm(range(1, 50)):
    dt = DecisionTreeClassifier(max_depth=d)
    dt_results[0].append(d)
    cv_scores = cross_val_score(dt, X_train, y_train, cv=10, scoring=accuracy)
    dt_results[1].append(cv_scores.mean())
    std_err = cv_scores.std() / np.sqrt(len(cv_scores))
    confidence_interval = 1.96 * std_err  # 95% доверительный интервал
    confidence_intervals.append(confidence_interval)

dt_results[0] = np.array(dt_results[0])
dt_results[1] = np.array(dt_results[1])
confidence_intervals = np.array(confidence_intervals)

plt.figure(figsize=(10, 6))
plt.plot(dt_results[0], dt_results[1], label='Mean accuracy', marker='+', color='green')
plt.fill_between(dt_results[0], 
                 dt_results[1] - confidence_intervals, 
                 dt_results[1] + confidence_intervals, 
                 color='green', alpha=0.2, label='95% confidence interval')
plt.xlabel("Depth of tree")
plt.ylabel("Accuracy on cross-validation")
plt.xticks(np.arange(1, 50, 2))
plt.title("Accuracy on Decision Tree")
plt.legend()
plt.grid()
plt.show()
```

```python
print("knn best accuracy:", max(knn_results[1]))
print("decision tree best accuracy:", max(dt_results[1]))
```

```python
best_knn = KNeighborsClassifier(n_neighbors=21)
best_tree = DecisionTreeClassifier(max_depth=7)
```

#### Алгоритмы справились примерно одинаково по качеству, но решающее дерево было всё-таки лучше. Также Дерево сработало быстрее. Возможно из-за более быстрого инференса О(depth) против O(n) у knn. Несмотря на то что knn обучается быстрее, за О(1). Но разница не очень большая, в целом результаты по времени сопоставимые

## PCA преобразование

Теперь попробуем сделать преобразрование данных с помощью PCA. Выберите оптимальное количество компонент и преобразуйте данные. Объясните свой выбор.

```python
# your code here
from sklearn.decomposition import PCA
pca = PCA()
plt.bar(np.arange(X_train.shape[1]), pca.fit(X_train).explained_variance_ratio_)
```

```python
print(f"Первые 20 компонент объясняют {np.sum(pca.fit(X_train).explained_variance_ratio_[:20]) * 100}% дисперсии")
```

#### Выберу 20 компонент, так как они объясняют 90% дисперсии (просто выбрал такой порог)

```python
X_train_PCA = PCA(n_components=20).fit_transform(X_train)
X_test_PCA = PCA(n_components=20).fit_transform(X_test)
X_train_PCA.shape, X_test_PCA.shape
```

## Обучение классификаторов на данных, трансформированных с помощью PCA

Обучим те же классификаторы на трансформированных с помощью PCA данных. Подберите оптимальные значения исследуемых гиперпараметров для новых данных. Постройте график среднего значения качества по кросс-валидации алгоритма при заданном значении гиперпараметра, на котором также отобразите доверительный интервал.

Что вы можете сказать о получившихся графиках? Какой алгоритм справился лучше? Какой быстрее? Почему?

```python
# your code here
accuracy = make_scorer(accuracy_score)

confidence_intervals = []
knn_results = [[],[]]
for n in tqdm(range(1, 100, 2)):
    neigh = KNeighborsClassifier(n_neighbors=n)
    knn_results[0].append(n)
    cv_scores = cross_val_score(neigh, X_train_PCA, y_train, cv=10, scoring=accuracy)
    knn_results[1].append(cv_scores.mean())
    std_err = cv_scores.std() / np.sqrt(len(cv_scores))
    confidence_interval = 1.96 * std_err  # 95% доверительный интервал
    confidence_intervals.append(confidence_interval)

knn_results[0] = np.array(knn_results[0])
knn_results[1] = np.array(knn_results[1])
confidence_intervals = np.array(confidence_intervals)

plt.figure(figsize=(10, 6))
plt.plot(knn_results[0], knn_results[1], label='Mean accuracy', marker='+', color='blue')
plt.fill_between(knn_results[0], 
                 knn_results[1] - confidence_intervals, 
                 knn_results[1] + confidence_intervals, 
                 color='blue', alpha=0.2, label='95% confidence interval')
plt.xlabel("Number of neighbors")
plt.ylabel("Accuracy on cross-validation")
plt.xticks(np.arange(1, 100, 4))
plt.title("Accuracy on KNN with PCA 20 components")
plt.legend()
plt.grid()
plt.show()
```

```python
accuracy = make_scorer(accuracy_score)

confidence_intervals = []
dt_results = [[],[]]
for d in tqdm(range(1, 50)):
    dt = DecisionTreeClassifier(max_depth=d)
    dt_results[0].append(d)
    cv_scores = cross_val_score(dt, X_train_PCA, y_train, cv=10, scoring=accuracy)
    dt_results[1].append(cv_scores.mean())
    std_err = cv_scores.std() / np.sqrt(len(cv_scores))
    confidence_interval = 1.96 * std_err  # 95% доверительный интервал
    confidence_intervals.append(confidence_interval)

dt_results[0] = np.array(dt_results[0])
dt_results[1] = np.array(dt_results[1])
confidence_intervals = np.array(confidence_intervals)

plt.figure(figsize=(10, 6))
plt.plot(dt_results[0], dt_results[1], label='Mean accuracy', marker='+', color='blue')
plt.fill_between(dt_results[0], 
                 dt_results[1] - confidence_intervals, 
                 dt_results[1] + confidence_intervals, 
                 color='blue', alpha=0.2, label='95% confidence interval')
plt.xlabel("Depth of tree")
plt.ylabel("Accuracy on cross-validation")
plt.xticks(np.arange(1, 50, 2))
plt.title("Accuracy on Decision Tree with PCA 20 components")
plt.legend()
plt.grid()
plt.show()
```

```python
best_pca_knn = KNeighborsClassifier(n_neighbors=11)
best_pca_tree = DecisionTreeClassifier(max_depth=5)
```
