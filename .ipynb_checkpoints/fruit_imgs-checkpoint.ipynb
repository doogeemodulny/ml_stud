```python
import matplotlib.pyplot as plt
```

```python
import torch
import torch.nn as nn
from torchvision.transforms import v2
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from torch.utils.data import random_split
```

```python
transformation = v2.Compose([v2.ToImage(), v2.ToDtype(torch.float32, scale=True)])
imagefolder = ImageFolder(root="fruit/", transform=transformation)
```

```python
classes = imagefolder.find_classes("fruit/")[0]
classes
```

```python
n = len(imagefolder.samples)
```

```python
generator = torch.Generator().manual_seed(2)
train_dataset, test_dataset = torch.utils.data.random_split(imagefolder, [0.8, 0.2], generator=generator)
len(train_dataset.indices), len(test_dataset.indices)
```

```python
torch.manual_seed(1)
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=64, shuffle=True)
```

```python
train_features, train_labels = next(iter(train_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
img = train_features[0].permute(1, 2, 0).squeeze()  #   
label = train_labels[0]
plt.imshow(img)
plt.show()
print(f"Label: {classes[label]}")
```

```python
image_shape = train_features.size()[2:]
image_shape
```

```python
def create_model(hid_size, num_features, n_out):
    torch.manual_seed(1)
    model = nn.Sequential(
        nn.Flatten(),
        nn.Linear(num_features, hid_size),
        nn.ReLU(),
        nn.Linear(hid_size, hid_size),
        nn.ReLU(),
        nn.Linear(hid_size, n_out),
        nn.Softmax(dim=1)
    )
    return model
```

```python
model = create_model(50, 3*image_shape[0]*image_shape[1], 15)
```

```python
torch.manual_seed(1)
test_features, test_labels = next(iter(test_dataloader))
print(test_features[0].shape)
output = model(test_features[0].unsqueeze(0))
print(output)
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from tqdm import tqdm

def train_model(model, opt, train_loader, test_loader, n_epoch=5):
    torch.manual_seed(1)  
    loss_fn = nn.CrossEntropyLoss() 
    acc_train_list, acc_test_list = [], []

    for epoch_num in range(n_epoch):
        train_loss, test_loss = 0.0, 0.0
        correct_train, correct_test = 0, 0
        total_train, total_test = 0, 0

        model.train()
        for images, labels in tqdm(train_loader):
            opt.zero_grad()
            y_predicted = model(images)
            loss = loss_fn(y_predicted, labels)

            loss.backward()
            opt.step()

            train_loss += loss.item()
            total_train += labels.size(0)
            _, predicted = torch.max(y_predicted, 1)
            correct_train += (predicted == labels).sum().item()

        train_acc = correct_train / total_train
        train_loss = train_loss / total_train
        acc_train_list.append(train_acc)

        model.eval()
        with torch.no_grad():
            for images, labels in test_loader:
                outputs = model(images)
                loss = loss_fn(outputs, labels)

                test_loss += loss.item()
                total_test += labels.size(0)
                _, predicted = torch.max(outputs, 1)
                correct_test += (predicted == labels).sum().item()

        test_acc = correct_test / total_test
        test_loss = test_loss / total_test
        acc_test_list.append(test_acc)

        print(f"Epoch {epoch_num + 1}/{n_epoch} " \
              f"Train Loss: {train_loss:.3f} - Train Acc: {train_acc:.3f} " \
              f"Test Loss: {test_loss:.3f} - Test Acc: {test_acc:.3f}")

    return model, acc_train_list, acc_test_list
```

```python
model = create_model(50, 3*image_shape[0]*image_shape[1], 15)
opt = torch.optim.Adam(params=model.parameters(), lr=0.001)
model, acc_train_list, acc_test_list = train_model(model, opt, train_dataloader, test_dataloader, n_epoch=5)
```

```python
plt.plot(acc_train_list, label='Train Accuracy')
plt.plot(acc_test_list, label='Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy during Training')
plt.show()

torch.manual_seed(1)  
test_features, test_labels = next(iter(test_dataloader))
with torch.no_grad():
    output = model(test_features)
    predicted_class = torch.argmax(output, dim=1) 
    predicted_value = output[0, predicted_class[0]].item()  
    print(f"Predicted class: {predicted_class[0].item()}")
    print(f"Predicted value: {predicted_value}")
    print(f"True label: {test_labels[0].item()}")
```
